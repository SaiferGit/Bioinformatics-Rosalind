#### BA3A: Generate the k-mer Composition of a String

```python

def CompositionK(string, k):
  kmers = set()
  for i in range(len(string)-k+1):
    kmers.add(string[i:i+k])
  return kmers

  

# main function
k = 5
string = "CAATCCAAC"
kmers = CompositionK(string, k)
for kmer in kmers:
  print(kmer)
```

#### BA3B: Reconstruct a String from its Genome Path

```python
def GenomeReconstruction(kmers):
  k = len(kmers[0])
  genome = kmers[0][:k-1]
  for kmer in kmers:
    genome += kmer[-1]
  return genome

# main function
with open ('/content/drive/MyDrive/Bioinformatics Lab/Rosalind Dataset/Lab Task 4/rosalind_ba3b.txt', 'r') as f:
    inputLines = [line.strip() for line in f.readlines()]

result = GenomeReconstruction(inputLines)
result
```

#### BA3C: Construct the Overlap Graph of a Collection of k-mers

```python
def getSuffix(string):
    return string[1:len(string)]

def getPrefix(string):
    return string[0:len(string)-1]

def ConstructOverlapGraph(genomelist):
  genomelist.sort()
  for s in genomelist:
    for p in genomelist:
      if getSuffix(s) == getPrefix(p):
        print(s +" -> "+p)

# main function
with open ('/content/drive/MyDrive/Bioinformatics Lab/Rosalind Dataset/Assignment/Chapter 3/rosalind_ba3c.txt', 'r') as f:
    inputLines = [line.strip() for line in f.readlines()]
ConstructOverlapGraph(inputLines)
```

#### BA3D: Construct the De Bruijn Graph of a String

```python
def getSuffix(string):
    return string[1:len(string)]

def getPrefix(string):
    return string[0:len(string)-1]

def DeBruijnK(genome, k):
  result = {}
  kmers = CompositionK(genome, k)
  for kmer in kmers:
    if getPrefix(kmer) not in result:
      result[getPrefix(kmer)] = getSuffix(kmer)
    else:
      result[getPrefix(kmer)] += ',' + getSuffix(kmer)
  return result

# main function
k = 4
genome = "AAGATTCTCTAC"

result = DeBruijnK(genome, k)

for key, value in sorted(result.items()):
  print(key + " -> " + value)
```

#### BA3F: Find an Eulerian Cycle in a Graph

```python
from random import choice

def RemoveEdge(adjList, startNode, targetNode):
  # first remove the edge from start to target
  adjList[startNode].remove(targetNode)
  if not adjList[startNode]: # if there is not any associated value with startNode 
    del adjList[startNode] # delete it from list
  return adjList

def FindEuerianCycle(adjList):
  # step 1: form a cycle Cycle by randomly walking in Graph (don't visit the same edge twice!)
  
  # randomly picking node and corresponding edges
  startNode, edges = choice(list(adjList.items()))

  # randomly picking one of the target nodes from edgelist
  targetNode = choice(edges)

  # as we have visited one edge, we can remove it from our adjList
  adjList = RemoveEdge(adjList, startNode, targetNode) 

  # our Cycle
  cycle = [startNode, targetNode]
  
  # making target node the current node from where we will continue the traversal
  currentNode = targetNode

  while currentNode != startNode:
    # find current node edges
    edges = adjList[currentNode]

    # picking a target node randomly
    targetNode = choice(edges)

    # now again we have visited currentNode -> targetNode, so we can remove this edge
    adjList = RemoveEdge(adjList, currentNode, targetNode)

    # updating currentNode
    currentNode = targetNode

    # appending new node into cycle
    cycle.append(currentNode)
  
  # now we have finished step 1. Our all adjList have to be removed. 
  # now we need to check for any unexpected edges
  # if any, then do the same process starting from unexpected edges
  # otherwise return the cycle
  while adjList:
    
    # checking if any node in cycle exist in adjList, storing them into potentialStarts
    potentiaStarts = [(idx, node) for idx, node in enumerate(cycle) if node in adjList]
        
    # choosing randomly a startpoint
    idx, newStart = choice(potentiaStarts)

    # forming new cycle starting from new start and then randomly walking
    newCycle = cycle[idx:] + cycle[1:idx+1]

    edges = adjList[newStart]
    
    # selecting new target node
    targetNode = choice(edges)

    adjList = RemoveEdge(adjList, newStart, targetNode)

    currentNode = targetNode
    newCycle.append(currentNode)

    # starting randomly traversal until currentNode != newStart
    while currentNode != newStart:
      edges = adjList[currentNode]
      targetNode = choice(edges)
      adjList = RemoveEdge(adjList, currentNode, targetNode)
      currentNode = targetNode
      newCycle.append(currentNode)
    cycle = newCycle

  return cycle

def ParseAdjList(adj_list_text):
  adjList = {}
  for line in inputLines:
    # split ->
    temp = line.split(' -> ')
    adjList[temp[0]] = temp[1].split(',')
  return adjList

# main function
with open ('/content/drive/MyDrive/Bioinformatics Lab/Rosalind Dataset/Assignment/Chapter 3/rosalind_ba3f.txt', 'r') as f:
    inputLines = [line.strip() for line in f.readlines()]


adjList = ParseAdjList(inputLines)
print(adjList)
print("->".join(FindEuerianCycle(adjList)))
```

#### BA3G: Find an Eulerian Path in a Graph

```python

def FindEuerianPath(adjList):

  # calculating digree difference i.e. out - in
  degdiff = {}
  for node, edges in adjList.items():
    # storing indegree
    if node in degdiff:
      degdiff[node] += len(edges)
    else:
      degdiff[node] = len(edges)

    # finding out outdegree of corresponding edges, if they exist in degdiff, that means
    # they have indegree assigned by previous if else. otherwise they have no indegree.
    # as #indegree is 1 in all the cases, so we can subtract 1 or assign -1
    for edge in edges:
      if edge in degdiff:
        degdiff[edge] -= 1
      else:
        degdiff[edge] = -1
  print(degdiff)
  # storing endingNodes
  for node, diff in degdiff.items():
    if diff == -1:
      endingNode = node

  # storing startingNodes
  for node, diff in degdiff.items():
    if diff == 1:
      startingNode = node
      

  # Euler path will exist iff euler cycle is exist
  #if ending node present draw edge from ending to starting node
  if endingNode in adjList:
    adjList[endingNode].append(startingNode)
  # else create a new edge from ending node to starting node
  else:
    adjList[endingNode] = startingNode
  print(adjList)
  eulerCycle = FindEuerianCycle(adjList)

  # let's form eulerpath from starting node to ending node
  # finding startingNode and endingNode index
  idx = 0
  while True:
    if eulerCycle[idx] == endingNode and eulerCycle[idx+1] == startingNode:
      break
    idx += 1
  eulerPath = eulerCycle[idx+1:] + eulerCycle[1:idx+1]
  return eulerPath

# main function
with open ('/content/drive/MyDrive/Bioinformatics Lab/Rosalind Dataset/Assignment/Chapter 3/rosalind_ba3g.txt', 'r') as f:
  inputLines = [line.strip() for line in f.readlines()]

adjList = ParseAdjList(inputLines)
print("->".join(FindEuerianPath(adjList)))
```

#### BA3H: Reconstruct a String from its k-mer Composition

```python
def getSuffixString(string):
  return string[1:]

def getPrefixString(string):
  return string[:-1]

def ReconstructGenomefromKmer(genomelist, k):
  suffix = []
  prefix = []
  dect = {} # for storing last position
  for genome in genomelist:
    suffix.append(genome[1:])
    prefix.append(genome[:-1])
    dect[genome[:k-1]] = genome[k-1:k]
  
  # finding starting genome
  for p in prefix:
    if p not in suffix:
      start = p
      break
  startgenome = start + dect[start]
  # finding ending pos
  for s in suffix:
    if s not in prefix:
      end = s
      break
  resultantPath = startgenome # initially
  
  while True:
    suffixPortion = resultantPath[len(resultantPath)-k+1: len(resultantPath)] # detemining current path's suffix portion
    if suffixPortion == end: # if it's equal to end then break
      break
    else: # else add the last char into path
      resultantPath += dect[suffixPortion]
  return resultantPath
    

  

# main
k = 4
with open('/content/drive/MyDrive/Bioinformatics Lab/Rosalind Dataset/Lab Task 5/rosalind_ba3h.txt', 'r') as f:
  genomelist = [line.strip() for line in f.readlines()]
path = ReconstructGenomefromKmer(genomelist, k)
path
```

#### BA3L: Construct a String Spelled by a Gapped Genome Path

```python


def StringSpelledByGappedGenomePatterns(gappedPatterns, k, d):
  prefix = suffix = ''
  
  # now traverse the gappedPatterns to construct prefix and suffix
  for i, patterns in enumerate(gappedPatterns):
    if i != len(gappedPatterns)-1:
      prefix += patterns[0][0]
      suffix += patterns[1][0]
    else:
      prefix += patterns[0]
      suffix += patterns[1]
  
  # now check whether the overlap is perfect or not
  for i in range(k+d+1, len(prefix)):
    if prefix[i] != suffix[i-k-d-1]:
      return "String can't be contructed"
  return prefix + suffix[len(suffix)-k-d-1: ]


# main function
k = 50
d = 200
with open ('/content/drive/MyDrive/Bioinformatics Lab/Rosalind Dataset/Assignment/Chapter 3/rosalind_ba3l.txt', 'r') as f:
  inputLines = [line.strip() for line in f.readlines()]

# generating reads from pair reads
gappedPatterns = []
for line in inputLines:
  gappedPatterns.append(line.split('|'))
print(StringSpelledByGappedGenomePatterns(gappedPatterns, k-1, d))
```
